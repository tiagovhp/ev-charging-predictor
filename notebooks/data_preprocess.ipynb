# Data Preprocess

```python
# Libraries
import pandas as pd
```

```python
# Read and import data
data = pd.read_csv("./data/processed/data_enriched.csv")

# Overview of data
display(data.head())
display(data.info())
```

```python
# Reorder columns by similar categories
# Define the ordered list of column names
ordered_columns = [
    'LOCATIONID', 'LATITUDE', 'LONGITUDE', 'TOWN', 'DISPLAYNAME', 'TOWN_NR', 'COUNTRY',
    'WEEKDAY', 'HOUR',
    'CONNECTOR_TYPES', 'NR_CONNECTORS', 'MIN_MAXELECTRIC_POWER_W', 'MAX_MAXELECTRIC_POWER_W',
    'CHARGING_CAPACITY', 'CHARGING_CAPACITY_2', 'CHARGE_PERCENTAGE', 'TOTAL_DURATION', 'AVG_DURATION',
    'parking', 'charging_station', 'fast_food', 'cafe', 'toilets', 'restaurant', 'convenience',
    'playground', 'park', 'community_centre', 'bar', 'pub', 'supermarket', 'sports_centre', 'hotel',
    'museum', 'theatre', 'hospital', 'library', 'zoo', 'cinema', 'bakery',
    'motorway', 'trunk', 'primary', 'secondary',
    'POPULATION', 'POP_DENSITY', 'AGE_20', 'AGE_40', 'AGE_60', 'AGE_80', 'AGE_100',
    'AVG_ICOME', 'COST_OF_LIVING', 'NET_INCOME',
    'CARS_PER_100', 'EV_RATE', 'AVG_TRAFFIC_DUR'
]

# Reindex the DataFrame using the ordered columns
data = data[ordered_columns]

# Overview of data
display(data.head())
display(data.info())
```

```python
''' Columns to drop:
    1. LOCATIONID
    2. TOWN
    2. DISPLAYNAME
    3. CONNECTOR_TYPES
    4. CHARGING_CAPACITY_2 (REDUNDANT)
    5. TOWN_NR
    6. COUNTRY
    7. AVG_DURATION
    6.
    7
'''
# Drop columns from dataset
data = data.drop(['LOCATIONID','TOWN', 'DISPLAYNAME', 'CONNECTOR_TYPES', 'TOWN_NR' ,'COUNTRY','AVG_DURATION','CHARGING_CAPACITY_2'],axis=1)
```

# Feature Engineering

```python
# Group 'HOUR' into categories / slots
'''
    6-10: Morning
    10-4: Midday
    4-8: Evening
    8-6: Night
'''
# Create a function to return time slot of a given hour
def time_slot(hour):
    if hour >= 6 and hour < 10:
        return "MORNING"
    elif hour >=10 and hour < 16:
        return "MIDDAY"
    elif hour >=16 and hour < 20:
        return "EVENING"
    else:
        return "NIGHT"

# Get time slots of our data['HOUR']
## list to hold time slots
slots = []
for hour in data['HOUR']:
    slot = time_slot(hour)
    slots.append(slot)

# Add a new feature (TIME_SLOT) and drop HOUR
data['TIME_SLOT'] = slots
data.drop(['HOUR'],axis=1,inplace=True)

# One hot encode time_slots
data = pd.get_dummies(data, columns=['TIME_SLOT'],prefix="")

# Convert Boolean to integers (if needed)
data[['_EVENING','_MIDDAY','_MORNING','_NIGHT']] = data[['_EVENING','_MIDDAY','_MORNING','_NIGHT']].astype(int)
```

```python
# Group 'Weekday' into Groups
'''
    0,1,2,3,4 : Workday
    5,6: Weekend
'''
# Function to return day type of a weekday
def day_type(weekday):
    if weekday <= 4:
        return "WORKDAY"
    else:
        return "WEEKEND"

# Get the day type of hour data['WEEKDAY']
# List to hold day type
type= []
for day in data['WEEKDAY']:
    type_day=day_type(day)
    type.append(type_day)

# Add a new feature ['DAY_TYPE'] and drop ['WEEKDAY']
data['DAY_TYPE'] = type
data.drop(['WEEKDAY'],axis=1,inplace=True)

# One hot encode time_slots
data = pd.get_dummies(data, columns=['DAY_TYPE'], prefix='')

# Convert Boolean to integers (if needed)
data[['_WORKDAY','_WEEKEND']] = data[['_WORKDAY','_WEEKEND']].astype(int)
```

```python
# Group amenities into Broader Groups
''' 
Food & Beverage: fast_food, cafe, restaurant, bar, pub, bakery
Retail & Convenience: supermarket, convenience
Competitors: charging_station
Leisure & Entertainment: playground, park, community_centre, sports_centre, museum, theatre, cinema, zoo
Health & Services: hospital, library, hotel, parking, toilets
'''

## Food & Beverage
data['FOOD_&_BEVERAGE'] = data['fast_food'] + data['cafe'] + data['restaurant'] + data['bar'] + data['pub'] + data['bakery']
## Retail & Convencience
data['RETAIL_&_CONVENIENCE'] = data['supermarket'] + data['convenience']
## Competitors
data['COMPETITORS'] = data['charging_station']
## Entretainment
data['ENTERTAINMENT'] = data['playground'] + data['park'] + data['community_centre'] + data['sports_centre'] + data['museum'] + data['theatre'] + data['cinema'] + data['zoo']
## Health & Services
data['HEALTH_&_SERVICES'] =  data['hospital'] + data['hotel'] + data['library'] + data['parking'] + data['toilets']


## Drop amenities columns
data = data.drop(['toilets','fast_food','cafe', 'restaurant', 'bar', 'pub' , 'bakery', 'supermarket','convenience', 'charging_station', 'playground', 'park', 'community_centre', 'sports_centre','museum','theatre','cinema','zoo', 'hospital','hotel','library','parking'],axis=1)
```

```python
# Set pandas to display all columns in .info() by setting the max_columns option
pd.set_option('display.max_columns', None)

# Now when you call .info(), all columns will be displayed
data.head()
#data.info()
```

```python
# List of columns to convert to float
columns_to_convert = ['CARS_PER_100', 'EV_RATE', 'POP_DENSITY', 'AGE_20', 'AGE_40', 'AGE_60', 'AGE_80', 'AGE_100']

# Replace commas with periods and convert to numeric
for col in columns_to_convert:
    data[col] = data[col].str.replace(',', '.', regex=False)  # Replace commas with periods
    data[col] = pd.to_numeric(data[col], errors='coerce')  # Convert to float, coerce errors to NaN

# Check the data types to confirm the conversion
print(data[columns_to_convert].dtypes)

display(data.head())
```

# Correlation between features

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.colors import LinearSegmentedColormap

# Create a custom colormap with transparency at zero
cmap = LinearSegmentedColormap.from_list(
    "custom_cmap", 
    [(0, 'red'), (0.5, 'white'), (1, 'red')],  # Red at both -1 and 1, white at 0
    N=256
)

# Correlation Matrix
corr_matrix = data.corr()

# Plot the correlation matrix
plt.figure(figsize=(80, 40))  # Increase the figure size
heatmap = sns.heatmap(
    corr_matrix, annot=True, fmt=".2f", cmap=cmap, square=True,
    vmin=-1, vmax=1, cbar_kws={'label': 'Correlation'}, annot_kws={'size': 20}
)

# Title for the plot
plt.title('Correlation Matrix')

# Set colorbar tick label font size
cbar = heatmap.collections[0].colorbar
cbar.ax.tick_params(labelsize=20)

# Set font size of x and y tick labels (feature names)
plt.xticks(fontsize=20)  # Adjust x-axis feature labels
plt.yticks(fontsize=20)  # Adjust y-axis feature labels

plt.show()
```

```python
# Drop features with high correlation
'''
AVG_INCOME - NET_INCOME
CHARGING_CAPACITY - NR_OF_CONNECTORS

 '''
data.drop(['AVG_ICOME','NR_CONNECTORS'], axis=1,inplace=True)

# We also drop TOTAL_DURATION because this is the same as the label to predict charge_percentage
data.drop(['TOTAL_DURATION'], axis=1,inplace=True)

data.head()
```
